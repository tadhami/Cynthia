```python
import itertools

elem = input("Enter a string: ")
lst = list(itertools.chain.from_iterable([["".join(p) for p in itertools.permutations(elem, l)] for l in range(1, len(elem)+1)]))
new_lst = []
for element in lst:
    item = sorted(set(element))
    if item not in new_lst:
        new_lst.append(item)
print(new_lst)
```
